```python
wsdl = 'http://agrovoc.uniroma2.it:8080/SKOSWS/services/SKOSWS?wsdl'
client = zeep.Client(wsdl=wsdl)
```

```python
chosen_language = "en"#"pt"
search_item = "pests"
terms_info = [term.replace('[', '').strip() for term in client.service.simpleSearchByMode2(searchString=search_item, searchmode="exact", separator=' ').split()[:-2]]
for i in range(len(terms_info)):
    if i % 3 == 0:
        print(client.service.getTermByLanguage(termcode=terms_info[i],language=chosen_language))
        print("Related items: ")
        related_items = [term.replace(']','').strip() for term in client.service.getConceptInfoByTermcode(termCode=terms_info[i])[-1].split(',')[1:]]
        for rel_item in related_items:
            name = client.service.getTermByLanguage(termcode=rel_item,language=chosen_language)
            if name != None:
                print(name)
        print("")
        print("Narrower items: ")
        narrow_items = [term.replace(']','').strip() for term in client.service.getConceptInfoByTermcode(termCode=terms_info[i])[-2].split(',')[1:]]
        for rel_item in narrow_items:
            name = client.service.getTermByLanguage(termcode=rel_item,language=chosen_language)
            if name != None:
                print(name)
        print("")
        print("Broader items: ")
        broader_names = [term.replace(']','').strip() for term in client.service.getConceptInfoByTermcode(termCode=terms_info[i])[-3].split(',')[1:]]
        for rel_item in broader_names:
            name = client.service.getTermByLanguage(termcode=rel_item,language=chosen_language)
            if name != None:
                print(name)
```

```python
import sys
sys.path.append('../src')

from utilities import excel_reader
from utilities import excel_writer
from time import sleep
import googletrans
from googletrans import Translator
import os
#from translate import Translator

def translate_keywords(file_to_read, file_translations, file_to_save):
    interventions_map = {}
    interv_df = excel_reader.ExcelReader().read_df_from_excel(file_to_read)
    if os.path.exists(file_translations):
        interv_translate_df = excel_reader.ExcelReader().read_df_from_excel(file_translations)
        for i in range(len(interv_translate_df)):
            if interv_translate_df["english word"].values[i].strip() not in interventions_map:
                interventions_map[interv_translate_df["english word"].values[i].strip()] = []
            interventions_map[interv_translate_df["english word"].values[i].strip()].append((
            interv_translate_df["foreign word"].values[i].strip(), interv_translate_df["language"].values[i].strip()))
    print("Found translations: ", len(interventions_map))
    
    translator = Translator()
    translations_interv = []
    all_translations_set = set()
    new_data = []
    cnt_translated = 0
    first_column = list(interv_df.columns)[0]
    print("Column to take: ", first_column)
    for i, keyword in enumerate(interv_df[first_column].values):
        all_langs = ["fr", "es", "pt"]
        langs_translated = []
        if keyword in interventions_map:
            new_data.append([keyword]+list(interv_df.values[i][1:]))
            for translation, language in interventions_map[keyword]:
                new_data.append([translation]+list(interv_df.values[i][1:]))
                langs_translated.append(language)
                if (translation, keyword, language) not in all_translations_set:
                    translations_interv.append((translation, keyword, language))
                    all_translations_set.add((translation, keyword, language))
        else:
            new_data.append([keyword]+list(interv_df.values[i][1:]))
            translations_interv.append((keyword, keyword, "en"))
            all_translations_set.add((keyword, keyword, "en"))
        
        langs_to_translate = set(all_langs) - set(langs_translated)
        if len(langs_to_translate):
            print(keyword)
            print(langs_to_translate)
            for lang in langs_to_translate:
                try:
                    res = translator.translate(keyword, src="en", dest=lang).text
                    #res = translator.translate(keyword)

                    if (res, keyword, lang) not in all_translations_set:
                        translations_interv.append((res, keyword, lang))
                        all_translations_set.add((res, keyword, lang))
                    new_data.append([res]+list(interv_df.values[i][1:]))
                    cnt_translated += 1
                except Exception as err:
                    print(err)
                    pass
            if cnt_translated % 50 == 0 and cnt_translated > 0:
                cnt_translated += 1
                sleep(5)
                translator = Translator()
    deduplicated_new_data = []
    deduplicated_new_data_set = set()
    for row in new_data:
        row = tuple(row)
        if row not in deduplicated_new_data_set:
            deduplicated_new_data_set.add(row)
            deduplicated_new_data.append(row)
    
    excel_writer.ExcelWriter().save_data_in_excel(translations_interv, ["foreign word", "english word", "language"], file_translations)
    excel_writer.ExcelWriter().save_data_in_excel(deduplicated_new_data, list(interv_df.columns), file_to_save)
```
